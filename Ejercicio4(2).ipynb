### 4. Clasificación Zero-Shot de Descripciones de Productos

Vamos a utilizar un modelo de clasificación zero-shot para categorizar las descripciones de nuestros productos existentes. Esto nos permitirá asignar una categoría basada en el texto descriptivo del producto, sin necesidad de entrenar un modelo específico para este dataset.

````python
from transformers import pipeline

# Inicializar el pipeline de clasificación zero-shot
# Usamos un un modelo de Hugging Face ampliamente utilizado y estable.
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=0 if torch.cuda.is_available() else -1)

# Definir las etiquetas candidatas para la clasificación
candidate_labels = ['laptop', 'monitor', 'perifericos', 'otros']

def classify_description_zero_shot(description):
    if pd.isna(description):
        return "otros"
    # Realizar la clasificación
    result = classifier(description, candidate_labels, multi_label=False)
    # Retornar la etiqueta con la puntuación más alta
    return result['labels'][0]

# Aplicar la clasificación a la columna 'descripcion' del DataFrame
df_products['categoria_zero_shot'] = df_products['descripcion'].apply(classify_description_zero_shot)

# Mostrar el DataFrame con la nueva columna de categoría zero-shot
display(df_products)
````

### 2. Módulo de Carga de Imágenes y Categorización
Este módulo permitirá al usuario subir una imagen. Un modelo de clasificación de imágenes de Hugging Face la categorizará, y el resultado se mostrará y se almacenará para el chatbot.

````python
import gradio as gr
from transformers import pipeline
from PIL import Image
import torch

# Define candidate labels for zero-shot image classification with more descriptive terms
image_candidate_labels = [
    'ordenador portatil, laptop',
    'pantalla de ordenador, monitor',
    'teclados, ratones o auriculares de ordenador (perifericos)',
    'elementos no relacionados con electronica de oficina (otros)'
]

# Cargar el pipeline de clasificación de imágenes zero-shot usando un modelo CLIP
image_classifier = pipeline(
    "zero-shot-image-classification",
    model="openai/clip-vit-large-patch14", # Updated model
    device=0 if torch.cuda.is_available() else -1
)

def classify_image(image: Image.Image):
    global predicted_image_category
    if image is None:
        predicted_image_category = None
        return "Por favor, sube una imagen para categorizar."

    # Realizar la clasificación zero-shot de la imagen con las etiquetas candidatas
    predictions = image_classifier(image, candidate_labels=image_candidate_labels)

    CONFIDENCE_THRESHOLD = 0.5 # Define a confidence threshold

    top_product_label = None
    top_product_score = 0.0
    other_score = 0.0

    for pred in predictions:
        full_label = pred["label"]
        score = pred["score"]

        # Directly identify the 'otros' label based on its unique phrase
        if 'otros)' in full_label: # 'elementos no relacionados con electronica de oficina (otros)'
            other_score = score
        else: # These are potential product categories
            # Map full label to simplified product category
            simplified_label = 'unknown'
            if 'laptop' in full_label:
                simplified_label = 'laptop'
            elif 'monitor' in full_label:
                simplified_label = 'monitor'
            elif 'perifericos' in full_label:
                simplified_label = 'perifericos'

            # Keep track of the highest scoring product category
            if simplified_label != 'unknown' and score > top_product_score:
                top_product_score = score
                top_product_label = simplified_label

    # Decision logic:
    # 1. If a product category has a score above threshold AND it's higher than 'otros' score
    # 2. Else if 'otros' score is above threshold (implies 'otros' is the most confident or product scores are low)
    # 3. Else fallback to 'otros' (no strong confidence in any specific category)
    if top_product_score >= CONFIDENCE_THRESHOLD and top_product_score > other_score:
        predicted_image_category = top_product_label
    elif other_score >= CONFIDENCE_THRESHOLD:
        predicted_image_category = 'otros'
    else:
        predicted_image_category = 'otros'

    if predicted_image_category == 'otros':
        return f"La imagen ha sido categorizada como: **{predicted_image_category.capitalize()}** (No se pudo categorizar con precisión entre los productos)."
    else:
        return f"La imagen ha sido categorizada como: **{predicted_image_category.capitalize()}**."
````

### 3. Chat de Información de Productos
Esta ventana de chat interactuará con el usuario. Si el usuario pregunta por la categoría de la imagen subida, el chatbot buscará y presentará los productos relevantes de nuestra base de datos.

````python
def format_products_as_table(category):
    filtered_products = [p for p in ecommerce_products if p["categoria"] == category]
    if not filtered_products:
        return "No se encontraron productos para esta categoría."

    table_header = "| Categoría | Modelo | Precio | Stock | Descripción |\n|---|---|---|---|---|\n"
    table_rows = []
    for p in filtered_products:
        table_rows.append(f"| {p['categoria'].capitalize()} | {p['modelo']} | {p['precio']}\u20ac | {p['stock']} | {p['descripcion']} |")

    return table_header + "\n".join(table_rows)

def get_product_models_for_category(category):
    """Returns a list of product model names for a given category."""
    filtered_products = [p for p in ecommerce_products if p["categoria"] == category]
    return [p["modelo"] for p in filtered_products]

def chat_response(message, history):
    global predicted_image_category # Access the global variable

    user_message_lower = message.lower()
    last_bot_message_lower = ""
    if history and history[-1][1]: # Check if there's a last bot message
        last_bot_message_lower = history[-1][1].lower()

    # Default values for product selection components (hidden and empty)
    product_choices = []
    product_selection_visible = False
    add_to_cart_visible = False

    response_text = "" # Initialize response_text

    # Scenario 1: User responded to the "Quieres consultar opciones?" prompt
    if "¿quieres consultar las opciones disponibles para la categoría?" in last_bot_message_lower:
        if any(keyword in user_message_lower for keyword in ["no", "no quiero", "gracias", "cancelar", "no me interesa"]):
            response_text = "Muy bien, puedes volver a subir otra imagen. Que tengas buen día!!!"
        elif any(keyword in user_message_lower for keyword in ["si", "sí", "quiero", "interesa", "por favor"]):
            if predicted_image_category and predicted_image_category != "No se pudo categorizar con precisión":
                product_choices = get_product_models_for_category(predicted_image_category)
                if product_choices:
                    response_text = f"Claro, aquí tienes los productos para la categoría '{predicted_image_category.capitalize()}'. Por favor, selecciona los que desees agregar al carrito a continuación:"
                    product_selection_visible = True
                    add_to_cart_visible = True
                else:
                    response_text = f"No se encontraron productos para la categoría '{predicted_image_category.capitalize()}' del producto categorizado."
            else:
                response_text = "Lo siento, no tengo una categoría de imagen reciente para consultar. Por favor, primero sube y categoriza una imagen."

        return response_text, product_choices, product_selection_visible, add_to_cart_visible

    # Scenario 2: User explicitly asks for image category info
    if "informacion" in user_message_lower and "categoria" in user_message_lower and "imagen" in user_message_lower:
        if predicted_image_category and predicted_image_category != "No se pudo categorizar con precisión":
            product_choices = get_product_models_for_category(predicted_image_category)
            if product_choices:
                response_text = f"Claro, aquí tienes los productos para la categoría '{predicted_image_category.capitalize()}'. Por favor, selecciona los que desees agregar al carrito a continuación:"
                product_selection_visible = True
                add_to_cart_visible = True
            else:
                response_text = f"No se encontraron productos para la categoría '{predicted_image_category.capitalize()}' del producto categorizado."
        else:
            response_text = "Lo siento, no tengo una categoría de imagen reciente para consultar. Por favor, primero sube y categoriza una imagen."

        return response_text, product_choices, product_selection_visible, add_to_cart_visible

    # Scenario 3: General Gemini response
    try:
        # Prepare chat history for Gemini in the required format
        gemini_history = []
        for user_msg, model_msg in history:
            if user_msg is not None:
                gemini_history.append({"role": "user", "parts": [user_msg]})
            if model_msg is not None:
                gemini_history.append({"role": "model", "parts": [model_msg]})
        gemini_history.append({"role": "user", "parts": [message]})

        response = gemini_model.generate_content(gemini_history)
        response_text = response.text
    except Exception as e:
        response_text = f"Lo siento, tuve un error al contactar a Gemini: {e}"

    return response_text, product_choices, product_selection_visible, add_to_cart_visible
````
